```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras

def plot_series(time,series,format='-',start = 0, end = None):
    plt.plot(time[start:end],series[start:end],format)
    plt.xlabel("Time")
    plt.ylabel("sales")
    plt.grid(True)
```

```python
import csv

time_step = []
sales = []
items = []
items_sales = []

with open('E:/Study/Masters/OpenCampus/OpenCampusProject/meteolytix_Corona_bakery_turnover_updated.csv') as csvfile:
    reader = csv.reader(csvfile,delimiter = ',')
    next(reader)
    step=0
    for row in reader:
        items.append(float(row[1]))
        sales.append(float(row[2]))
        time_step.append(step)
        step=step+1
        
series = np.array((items,sales))
time = np.array(time_step)
plt.figure(figsize=(20,10))



for y in range(1,len(np.unique(items))+1):
    val = []
    strt = 0
    for x in range(len(items)):
        if series[0][x] == y:
            val.append(float(series[1][x]))
    items_sales.append(val)
    end = len(val)
    

print(len(items_sales[0]))
print(len(items_sales[1]))
print(len(items_sales[2]))
print(len(items_sales[3]))
print(len(items_sales[4]))
print(len(items_sales[5]))

plot_series(time[0:len(items_sales[0])],items_sales[0])
plot_series(time[0:len(items_sales[5])],items_sales[5])
```
